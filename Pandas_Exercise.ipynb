```python
def intensity(x):
    if x <= 250:
        return 'Low'
    elif (x > 250) and (x <= 300):
        return 'Medium'
    else:
        return "High"
    
df["Intensity"] = df["Calories"].apply(intensity)
df
```

```python
df.head(10)
```

```python
df = df.sort_values("Duration")
df
```

```python
df["Duration"] = df.Duration.replace(450,45)
```

```python
df = df.sort_index()
df
```

```python
date = df[df["Date"].isna()]
date
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
```

```python
for index, row in date.iterrows():
    days_to_add = 1
    Previous_record = df.loc[index-1]
    df.loc[index,"Date"]= pd.to_datetime(Previous_record["Date"])+pd.to_timedelta(days_to_add,unit='d')
```
