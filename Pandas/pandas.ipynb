Series: a one-dimensional labeled array holding data of any type

DataFrame: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

GROUP BY

```python
data.groupby('Purchase_Category')['Purchase_Amount'].sum()

data.groupby(['Purchase_Category', 'Gender'])['Purchase_Amount'].mean()

df1 = data.groupby(['Purchase_Category', 'Location']).agg(
    TOTAL_PURCHASE_AMOUNT=('Purchase_Amount', 'sum'), 
    NUNIQUE_CUSTOMERS=('Customer_ID', 'nunique'),  
).reset_index()
```

```python
import pandas as pd

dt = {
    "Customer_ID": ["C001", "C002", "C003"],
    "Electronics": [200, 150, 180],
    "Clothing": [100, 250, 200],
    "Furniture": [300, 400, 500]
}

df = pd.DataFrame(dt)

df_stacked = df.set_index("Customer_ID").stack().reset_index()
df_stacked.columns = ["Customer_ID", "Category", "Purchase_Amount"]

print(df_stacked)
```

Pivot

--> .pivot(): Reshapes data using unique values in columns (fails if duplicate values exist).

--> .pivot_table(): Allows aggregations and works even if duplicate values exist.

```python
df_pivot_table = data.pivot_table(index="Customer_ID", columns="Purchase_Category", values="Purchase_Amount", aggfunc="sum")
print(df_pivot_table)
```

Merge and Concat

```python
data1 = pd.DataFrame({
    "Customer_ID": ["C001", "C002", "C003", "C004"],
    "Name": ["Alex", "Mia", "Ryan", "Zara"]
})
```

```python
data2 = pd.DataFrame({
    "Customer_ID": ["C001", "C002", "C005", "C006"],
    "Purchase_Amount": [200, 150, 300, 400]
})
```

```python
pd.merge(data1,data2)
```

```python
merged_df = data1.merge(data2, on="Customer_ID", how="inner")
print(merged_df)
```
