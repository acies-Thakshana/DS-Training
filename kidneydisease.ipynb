```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Convert object-type numerical columns to float
for col in ["pcv", "wc", "rc"]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing values
for col in df.select_dtypes(include=['number']).columns:
    df[col].fillna(df[col].median(), inplace=True)  # Median for numerical

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)  # Mode for categorical

# Convert categorical variables using Label Encoding
label_enc = LabelEncoder()
for col in ["rbc", "pc","pe", "pcc", "ba", "htn", "dm", "cad", "appet", "ane", "classification"]:
    df[col] = label_enc.fit_transform(df[col])
scaler = StandardScaler()
```
