```python
import pandas as pd
import plotly.express as pl
```

```python
data = pd.read_excel(r"C:\Users\DYNABOOK\Downloads\excel sheet.xlsx", sheet_name="Sales - Data")
```

```python
data['Order date'] = pd.to_datetime(data['Order Date'])
data
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

df_sales = data.groupby("Order date")["Sales"].sum().reset_index()
df_sales
df_countries = data.groupby("Order date")["Country"].nunique().reset_index()

fig = make_subplots(rows=2, cols=1, shared_xaxes=True,subplot_titles=("Line : Sales Sum", " Bar: Unique Countries"))


fig.add_trace(
    go.Scatter(x=df_sales["Order date"], y=df_sales["Sales"], fill='tozeroy',mode="lines", name="Sales"),
    row=1, col=1
)

fig.add_trace(
    go.Bar(x=df_countries["Order date"], y=df_countries["Country"], name="Unique Countries"),
    row=2, col=1
)

fig.update_layout(title="Sales and Country Analysis using Plotly", showlegend=True)

fig.show()
```

```python
df1 = data.groupby(['Order Date','Order Priority']).agg(
    SUM_SALES=('Sales','sum')
    , NUNIQUE_COUNTRY=('Country', 'nunique')
    , UNIQUE_COUNTRY_LIST=('Country', 'unique')
).reset_index()

df1['Nunique country from list'] = df1['UNIQUE_COUNTRY_LIST'].apply(lambda country_list: len(country_list))
fig = make_subplots(rows=2, cols=1, shared_xaxes=True)


fig.add_trace(
    go.Scatter(x=df_sales["Order date"], y=df_sales["Sales"], fill='tozeroy',mode="lines", name="Sales"),
    row=1, col=1
)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pivot = pd.pivot_table(data, 
                             values="Sales", 
                             index="Order Date", 
                             columns="Order Priority",
                             aggfunc="sum",
                             ).reset_index()

fig = make_subplots(rows=2, cols=1, shared_xaxes=True,subplot_titles=("Line : Sales Sum"))


for col in pivot.columns[1:]: 
    fig.add_trace(
        go.Scatter(x=pivot["Order Date"], y=pivot[col], fill='tozeroy', mode="lines", name=col),
        row=1, col=1
    )
fig.show()

```
