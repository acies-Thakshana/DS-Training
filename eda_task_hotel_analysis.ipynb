```python
#Univariate analysis

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(hotel_df['avg_price_per_room'], bins=30, kde=False, color='blue')
plt.title('Distribution of Room Prices')
plt.xlabel('Room Price')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8,5))
sns.countplot(data=hotel_df, x='type_of_meal_plan',hue='type_of_meal_plan', palette='coolwarm')
plt.title('Most Popular Meal Plan')
plt.xlabel('Meal Plan')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=hotel_df, x='room_type_reserved',hue='type_of_meal_plan', palette='viridis')
plt.title('Most Booked Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_price_per_room = hotel_df.groupby('room_type_reserved')['avg_price_per_room'].mean().reset_index()

# Plot the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x='room_type_reserved', y='avg_price_per_room', data=avg_price_per_room, palette="Blues_r")

# Add labels and title
plt.xlabel("Room Type")
plt.ylabel("Average Room Price ($)")
plt.title("Average Room Price Across Room Types")
plt.xticks(rotation=45)
plt.show()
```

```python

repeated_guests = hotel_df[hotel_df['repeated_guest'] == 1].groupby('room_type_reserved').size().reset_index(name='count')
repeated_guests['Guest Type'] = 'Repeated Guests'

special_guests = hotel_df[hotel_df['no_of_special_requests'] > 2].groupby('room_type_reserved').size().reset_index(name='count')
special_guests['Guest Type'] = 'Special Guests'


guest_analysis = pd.concat([repeated_guests, special_guests])

plt.figure(figsize=(8,5))
sns.barplot(data=guest_analysis, x='room_type_reserved', y='count', hue='Guest Type', palette='coolwarm')

plt.xlabel("Room Type Reserved")
plt.ylabel("Number of Guests")
plt.title("Room Type Preference of Repeated and Special Guests")
plt.xticks(rotation=45)
plt.legend(title="Guest Type")
plt.show()
```

```python
import pandas as pd
import plotly.express as px
hotel_df['Guest Type'] = hotel_df.apply(lambda x: 'Repeated Guest' if x['repeated_guest'] == 1 
                                        else ('Special Guest' if x['no_of_special_requests'] > 2 else 'New Guest'), axis=1)
sunburst_data = hotel_df.groupby(['Guest Type', 'room_type_reserved']).size().reset_index(name='count')
fig = px.sunburst(sunburst_data, 
                  path=['Guest Type', 'room_type_reserved'], 
                  values='count',
                  title="Room Type Preference of Repeated and Special Guests",
                  color='Guest Type', 
                  color_discrete_map={'Repeated Guest': 'blue', 'Special Guest': 'red', 'New Guest': 'green'})
fig.show()
```

```python
monthly_bookings = hotel_df.groupby(['arrival_year', 'arrival_month'])['Booking_ID'].count().reset_index()
pivot_data = monthly_bookings.pivot(index='arrival_month', columns='arrival_year', values='Booking_ID')
plt.figure(figsize=(12, 6))
for year in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[year], marker='o', linestyle='-', label=str(year))
plt.xticks(range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Monthly Booking Trend Across Years")
plt.legend(title="Year")
plt.grid(True)

plt.show()
```

```python

oct_nov_data = hotel_df[hotel_df['arrival_month'].isin([10, 11])]
plt.figure(figsize=(10, 5))
sns.countplot(data=oct_nov_data, x='market_segment_type', hue='arrival_month', palette='coolwarm')
plt.xlabel("Market Segment")
plt.ylabel("Number of Bookings")
plt.title("Market Segment Contribution - October vs November")
plt.legend(title="Month", labels=['October', 'November'])
plt.xticks(rotation=45)
plt.show()
```

```python
stay_analysis = oct_nov_data.groupby('arrival_month')[['no_of_adults', 'no_of_children']].sum().reset_index()

plt.figure(figsize=(8, 5))
stay_analysis.set_index('arrival_month').plot(kind='bar', stacked=True, color=['blue', 'orange'])

plt.xlabel("Month")
plt.ylabel("Total Count")
plt.title("Total Adults and Children Staying in October & November")
plt.legend(["Adults", "Children"])
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
january_bookings = hotel_df[hotel_df['arrival_month'] == 1]
def categorize_guests(row):
    if row['no_of_adults'] == 1 and row['no_of_children'] == 0:
        return 'Solo Traveler'
    elif row['no_of_adults'] == 2 and row['no_of_children'] == 0:
        return 'two people room' \
        ''
    elif row['no_of_adults'] >= 1 and row['no_of_children'] > 0:
        return 'Family'
    elif row['no_of_adults'] >= 3:
        return 'Group'
    else:
        return 'Other'  
january_bookings.loc[:, 'Guest Category'] = january_bookings.apply(categorize_guests, axis=1)
guest_analysis = january_bookings['Guest Category'].value_counts().reset_index()
guest_analysis.columns = ['Guest Category', 'Number of Bookings']

print(guest_analysis)
```

```python

meal_analysis = january_bookings.groupby(['Guest Category', 'type_of_meal_plan']).size().reset_index(name='count')
meal_pivot = meal_analysis.pivot(index='Guest Category', columns='type_of_meal_plan', values='count')
plt.figure(figsize=(8, 5))
sns.heatmap(meal_pivot, annot=True, cmap='coolwarm', fmt='g')

plt.xlabel("Type of Meal Plan")
plt.ylabel("Guest Category")
plt.title("Meal Preference by Guest Category in January")
plt.show()
```
