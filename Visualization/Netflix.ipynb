```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("netflix_users.csv")
```

```python
data.head()
```

```python
import plotly.express as px

df1 = data.groupby(['Country']).agg(
Count_Country=('User_ID','nunique')


).reset_index()


fig = px.bar(df1, x='Country',y='Count_Country')
fig.show()
```

```python
fig = px.pie(data, names='Subscription_Type', title="Netflix Subscription Types")
fig.show()
```

```python
fig = px.sunburst(data, path=['Subscription_Type', 'Favorite_Genre'], values='Watch_Time_Hours')
fig.show()
```

```python
avg_watch_time = data.groupby("Country")["Watch_Time_Hours"].mean().reset_index()

fig = px.bar(avg_watch_time, x='Country', y='Watch_Time_Hours', title="Average Watch Time per Country")
fig.show()
```

```python
df2 = data.groupby(['Age']).agg(
Sum_Hours=('Watch_Time_Hours','sum')


).reset_index()
fig = px.scatter(df2,x='Age', y='Sum_Hours')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

data["Last_Login"] = pd.to_datetime(data["Last_Login"])  # Convert to datetime
watch_time_trend = data.groupby(data["Last_Login"].dt.date)["Watch_Time_Hours"].sum().reset_index()

fig = px.line(watch_time_trend, x="Last_Login", y="Watch_Time_Hours", title="Watch Time Trend Over Time")
fig.show()
```

```python
import plotly.express as px

fig = px.box(data, x="Subscription_Type", y="Watch_Time_Hours", title="Watch Time by Subscription Type")
fig.show()
```

```python
data['Age'].max()
```

```python
bins = [18, 25, 30, 35, 40, 50, 60, 70, 80]
labels = ["18-25", "26-30", "31-35", "36-40", "41-50", "51-60", "61-70", "71-80"]

# Create a new column for Age Group
data["Age_Group"] = pd.cut(data["Age"], bins=bins, labels=labels, right=True)

# Aggregate count of users in each Age Group per Country
df = data.groupby(["Country", "Age_Group"]).size().reset_index(name="Count")

# Create a grouped bar chart using the modified `data`
fig = px.bar(df, x="Country", y="Count", color="Age_Group", 
             barmode="group", title="Grouped Bar Chart of Age Group by Country",
             color_discrete_sequence=["blue", "lightblue", "darkblue", "royalblue", "skyblue"])

# Update background color
fig.update_layout(
    plot_bgcolor="white",   
    paper_bgcolor="black",      
    font=dict(color="white")    
)

# Show the plot
fig.show()
```

```python
import plotly.express as px

filtered_data = data.query("Age_Group == '18-25'")  

fig = px.bar(
    filtered_data,
    x="Favorite_Genre",
    y="Watch_Time_Hours",
    title="Total Watch Time by Genre (Ages 18-25)",
    labels={"Watch_Time_Hours": "Total Watch Time (Hours)", "Favorite_Genre": "Genre"},
    barmode="group"
)

fig.show()
```

```python
romance_data = data.query("Favorite_Genre == 'Romance'")

romance_watch_summary = romance_data.groupby("Subscription_Type")["Watch_Time_Hours"].sum().reset_index()

romance_watch_summary = romance_watch_summary.sort_values(by="Watch_Time_Hours", ascending=False)

print(romance_watch_summary)
```
