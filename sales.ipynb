```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv(r"C:\Users\DYNABOOK\Downloads\superstore_sales.csv")
```

```python

fig = px.treemap(data, 
                 path=['Region', 'Segment', 'Customer Name'],  
                 values='Sales',  
                 color='Sales',  
                 title="Customer Distribution by Region & Segment")


fig.show()
```

```python

fig = px.choropleth(data, 
                    locations="State",  
                    locationmode="USA-states",  
                    color="Sales",  
                    title="Sales Distribution Across States",
                    scope="usa") 


fig.show()
```

```python
top_state = "TX"  # Replace with actual top state from analysis
state_data = data[data["State"] == top_state]

fig = px.bar(state_data, x="Sub-Category", y="Sales", color="Product Category", 
             title=f"Top Selling Product Categories in {top_state}", 
             labels={"Sales": "Total Sales", "Sub-Category": "Product Sub-Category"},
             barmode="group")
fig.show()
```

```python
top_product = "Accessories"  # Replace with actual best-seller
product_data = state_data[state_data["Product Name"] == top_product]

fig = px.pie(product_data, names="Customer Name", values="Sales", 
             title=f"Top Salespersons for {top_product} in {top_state}")
fig.show()
```

```python
fig = px.pie(state_data[state_data["Product Name"] == "Accessories"], 
             names="Customer Name", values="Sales", 
             title=f"Top Customers for Accessories in {top_state}")  
fig.show()
```

```python
top_n = 10  # Change this number if needed
top_products = state_data.groupby("Product Name")["Profit"].sum().nlargest(top_n).reset_index()

fig = px.bar(top_products, x="Product Name", y="Profit", 
             title=f"Top {top_n} Profitable Products in {top_state}", 
             labels={"Profit": "Profit", "Product Name": "Product"})
fig.show()
```
