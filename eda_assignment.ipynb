```python
df['booking_status']= df['booking_status'].replace({'Not_Canceled':1,'Canceled':0})
```

```python
cancelled_count = df['booking_status'].value_counts(normalize = 'True')
```

```python
cancelled_count
```

```python
import seaborn as sns
sns.boxplot(x=df["booking_status"], y=df["lead_time"])
plt.title("Lead Time vs Booking Status")
plt.xlabel("Booking Status")
plt.ylabel("Lead Time")
```

Canceled bookings tend to have a higher lead time. so customers who book far in advance are more likely to cancel.


```python
sns.countplot(x=df["arrival_month"], hue=df["booking_status"], palette="coolwarm")
plt.title("Monthly Booking Trends")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend(title="Booking Status")
```

More cancellations in months like october. Decemeber cancellation is less compared to bookings done in that month

```python
sns.boxplot(x=df["booking_status"], y=df["no_of_special_requests"])
plt.title("Special Requests vs Booking Status")
plt.xlabel("Booking Status")
plt.ylabel("Number of Special Requests")
```

 special requests do not seem to strongly impact whether a booking gets canceled or not.

```python
sns.countplot(x=df["required_car_parking_space"], hue=df["booking_status"], palette="muted")
plt.title("Parking Requirement vs Booking Status")
plt.xlabel("Car Parking Space Required")
plt.ylabel("Count")
plt.legend(title="Booking Status")
```

Cancellations are more common for bookings without parking, but the difference is not drastic.
Guests who require parking are less likely to cancel.

```python
import plotly.express as px
fig = px.histogram(df, x="market_segment_type", color="booking_status", barmode="group", 
                   title="Market Segment vs Cancellations", labels={"market_segment_type": "Market Segment", "count": "Count"},
                   color_discrete_map={"Canceled": "lightblue", "Not Canceled": "salmon"})
fig.show()
```

cancellations more in online due to flexible cancellation policies by apps

```python
plt.figure(figsize=(12, 6))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x=df["no_of_special_requests"], hue=df["required_car_parking_space"], palette="viridis")
plt.title("Special Requests vs Parking Requirement")
plt.xlabel("Number of Special Requests")
plt.ylabel("Count")
plt.legend(title="Car Parking Space Required")
```

number of special requests does not strongly influence whether parking is required or not.
